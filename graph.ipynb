```python
import datetime 

data =  read_file('election_tweets.csv') 
data = data[1:len(data)]  # get rid of column labels

# This function converts the dates to a standard format
def convert_date(date):
    vec = date.split('-')
    year = vec[0]
    month = vec[1]
    day = vec[2]
    #print int(month), int(year), int(day)
    date = datetime.date(int(year), int(month), int(day))
    # print date
    
    startdate = datetime.date(2016, 07, 10)
    enddate = datetime.date(2016, 11, 07)
    
    if date < startdate or date > enddate: #and int(day) < 10:
        return "continue"
    else:
        date = month+"/"+day+"/"+year
        return date
    
    
# counts the number of tweets tweeted per day
num = 0
for row in data:
    d = convert_date(row[0].split(' ')[0])  
    if d == "continue":
        continue
    else:
        num += 1
            
print "TOTAL Tweets between 2016-07-10 and 2016-11-07 ",  num
```

```python
# This function determines the label (ie. vote)
# 1: vote for clinton
# 0: vote for trump
def calculate_vote(row):
    if row['Converted_Clinton']>=row['Converted_Trump']:
        return 1
    elif row['Converted_Clinton']<row['Converted_Trump']:
        return 0
    
```

```python
############# Calculate Bias ############################
# This is the section where I see if simply counting the 'votes' is a better predictor than the linear svm

# function to calculate the bias -- poll dataset
def find_bias(row):
    APC1_f = 48.89
    APC1_s = row['Converted_Trump']
    return APC1_s - APC1_f


# function to calculate the bias  -- twitter dataset
def find_bias2(row):
    APC1_f = 48.89
    APC1_s = row['Percent_forTrump']
    return APC1_s - APC1_f
    
```

```python
# To calculate bias (only against Trump) Im using the equation
# BC1 = APC1,s - APC1,f
# the adjusted percentage of candidate 𝐶1 in
# the actual election and the adjusted percentage of candidate
# 𝐶1 in the poll currently being analyzed.
# equation from https://arxiv.org/ftp/arxiv/papers/1701/1701.06232.pdf
# ** BC1 = -BC2

# Figure 3

df = pd.read_csv("pop_2party.csv")
df['Bias_Against_Trump'] = df.apply(find_bias, axis=1)

df2 = pd.read_csv("election_tweets_counts6.csv")
df2['Bias_Against_Trump'] = df2.apply(find_bias2, axis=1)

trace = Scatter( x=df['Date'], y=df['Bias_Against_Trump'],
                marker=dict(color = "purple" ),
                name = 'Bias Against Trump (Polling Data)')

trace2 = Scatter( x=df2['Date'], y=df2['Bias_Against_Trump'],
                marker=dict(color = "orange" ),
                name = 'Bias Against Trump (Twitter Data)')

layout = Layout(
    title="Bias Against Trump by Day",
    plot_bgcolor = 'rgb(230, 230,230)',
    showlegend=True)

fig = Figure(data = [trace, trace2], layout=layout )
iplot(fig, filename="bias")
```
